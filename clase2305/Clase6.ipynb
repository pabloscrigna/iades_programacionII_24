#### Heterogeneos

En otros lenguajes, las colecciones tienen una restricción la cual solo permite tener un tipo de dato. Pero en Python, no tenemos esa restricción. Podemos tener un conjunto heterogéneo que contenga números, variables, strings, o tuplas.

```python
# Ejemplo heterogeneo
# set heterogeneo int - float - str
conjunt1 = {0, 4.6, "34"}
conjunt1
```

```python
# Ejemplo heterogeneo con variables
numero = 89
letra = "2"
peso = 7.8

conjunto = {numero, letra, peso}
conjunto
```

```python
# A partir de un iterable
# set(lista)
lista = [0,1,2,3,4,5,6]
set_lista = set(lista)
set_lista
```

```python
# Si paso esta lista a set, como queda el conjunto?
lista = [7,1,1,2,3,3]
set_lista = set(lista)
set_lista
```

```python
# tupla
tupla = (5,5,3,5,6,7, 7)

set_tupla = set(tupla)
set_tupla
```

```python
# Crear un set con range.
conjunto = set(range(10,20))
conjunto
```

```python
# Ejemplo con listas
conjunto_lista = {0, 1, [9,5]}
conjunto_lista
```

#### Conversiones

Un set puede ser convertido a una lista y viceversa. En este último caso, los elementos duplicados son unificados.


```python
# Convertir set a lista (list)

# partimos de un conjunto y llegamos a una lista

conjunto = {0, 1, 2}

conj_list = list(conjunto)
conj_list
```

```python
#convertir lista a set

lista = ["2", "3"]

conjunto = set(lista)
conjunto
```

## Diccionarios

Un diccionario **dict** es una colección no ordenada de objetos. Es por eso que para **identificar** un valor cualquiera dentro de él, especificamos una **clave** (a diferencia de las listas y tuplas, cuyos elementos se identifican por su posición).

Las claves suelen ser int o string, aunque cualquier otro objeto inmutable puede actuar como una clave. Los valores, por el contrario, pueden ser de cualquier tipo, incluso otros diccionarios.

```python
# diccionario vacio
diccionario = {}
diccionario
```

```python
type(diccionario)
```

```python
# con datos -- los pares clave:valor se separan por coma
# keys : nombre edad peso
persona = {
    "nombre" : "Juan",
    "edad" : 34,
    "peso" : 34.7
}
```

```python
print(persona)
```

```python
# acceder al campo peso?
persona["peso"]
```

```python
persona["nombre"]
```

```python
#diccionario con otro diccionario

alumno = {
    "nombre": "Juan",
    "curso": "50190",
    "caracteristicas": {
        "altura": 1.24,
        "peso": 45
    },
    "notas": [8,6,9]
}
```

```python
alumno
```

```python
# type
type(alumno)
```

```python
alumno["caracteristicas"]
```

```python
alumno["caracteristicas"]["altura"]
```

#### Leyendo los valores

```python
# nombre_dict[clave] --

persona = {
    "nombre" : "Juan",
    ("python", 50190): 9,
    "datos" : {
        "altura": 23,
        "peso": 50
    }
}

print(persona["nombre"])
print(persona[("python", 50190)])
print(persona["datos"])
print(persona["datos"]["peso"])
```
